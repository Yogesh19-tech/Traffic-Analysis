```python
import pandas as pd
import numpy as np
from faker import Faker
import random

# Initialize faker
fake = Faker()

# Set random seed for reproducibility
random.seed(42)
np.random.seed(42)

# Number of dummy records
n = 5000

# Generate dummy data
data = {
    "id": range(1, n+1),
    "Gender": [random.choice(["Male", "Female"]) for _ in range(n)],
    "Age": [random.randint(18, 75) for _ in range(n)],
    "Travel category": [random.choice(["Personal", "Business"]) for _ in range(n)],
    "Travel Class": [random.choice(["Economy", "Premium", "Business"]) for _ in range(n)],
    "Distance Travelled": [random.randint(20, 2000) for _ in range(n)],
    "Departure/Arrival Rating": [random.randint(1, 5) for _ in range(n)],
    "Booking Ease": [random.randint(1, 5) for _ in range(n)],
    "Boarding Point": [fake.city() for _ in range(n)],
    "Food": [random.randint(1, 5) for _ in range(n)],
    "Seat Comfort": [random.randint(1, 5) for _ in range(n)],
    "Entertainment": [random.randint(1, 5) for _ in range(n)],
    "Leg Room": [random.randint(1, 5) for _ in range(n)],
    "Luggage Handling": [random.randint(1, 5) for _ in range(n)],
    "Cleanliness": [random.randint(1, 5) for _ in range(n)],
    "Departure Delay (min)": [max(0, int(np.random.normal(15, 10))) for _ in range(n)],
    "Arrival Delay (min)": [max(0, int(np.random.normal(20, 15))) for _ in range(n)],
}

# Satisfaction outcome based on some logic (dummy)
satisfaction = []
for i in range(n):
    if data["Seat Comfort"][i] >= 4 and data["Cleanliness"][i] >= 4 and data["Departure Delay (min)"][i] < 30:
        satisfaction.append("satisfied")
    else:
        satisfaction.append(random.choice(["satisfied", "dissatisfied"]))
data["Satisfaction"] = satisfaction

# Create DataFrame
df_dummy = pd.DataFrame(data)


# Save to CSV for reuse
df_dummy.to_csv("dummy_transport_satisfaction.csv", index=False)

df_dummy.head()
```
